# 5.1 합성곱 신경망(CNN)

- 합성곱 신경망은 이미지 전체를 한 번에 계산하는 것이 아니라, 이미지의 국소적 특징을 작은 필터(커널)로 추출하는 방식으로 계산량을 줄이고 효율적으로 학습하는 신경망 구조이다.

### 5.1.1 합성곱층의 필요성

- 이미지를 납작하게(Flatten) 펼치면 공간 구조 정보를 잃게 된다.  
CNN은 이를 막기 위해 작은 커널을 이용해 이미지의 공간적 패턴을 보존한 채 특징을 추출한다.

### 5.1.2 합성곱 신경망 구조

CNN의 기본 계층 구성:
1) 입력층(Input layer)  
2) 합성곱층(Convolution layer)  
3) 풀링층(Pooling layer)  
4) 완전연결층(Fully-connected layer)  
5) 출력층(Output layer)

- 합성곱층과 풀링층을 반복하며 특징 벡터를 추출하고, 마지막 완전연결층에서 Softmax로 분류한다.

1) 입력층

- 이미지의 형태(shape)는  
(높이 H, 너비 W, 채널 수 C)  
- 예: RGB → (H, W, 3) / 그레이스케일 → (H, W, 1)

2) 합성곱층(Convolution Layer)

- 커널(필터)을 이미지에 이동시키며 곱셈-합산 연산을 수행해 특징을 추출  
- 커널 크기 예: 3×3, 5×5  
- 스트라이드(stride): 한 번에 이동하는 칸수  
- 결과는 특징 맵(feature map)

- 합성곱 연산 예시(3×3 필터, stride=1):  
- 입력 이미지의 각 3×3 영역 × 필터 값을 곱하고 모두 더해 출력 1칸 생성  
- 이 과정을 전체 이미지에 대해 반복 수행

- 여러 단계의 필터 이동 예시 요약 (stride=1)

1단계 → 첫 번째 3×3 영역과 필터 연산 → 결과 3  
2단계 → 오른쪽으로 한 칸 이동 → 결과 1  
3단계 → 다시 이동 → 결과 1  

최종 출력 특징맵 완성

- 컬러 이미지 합성곱

- RGB처럼 채널이 3개면 필터도 입력 채널과 동일한 3개의 채널을 가진다  
- 필터가 여러 개면 출력 채널 수가 증가  

예: 필터 2개 → 출력 특징맵도 2개의 채널 생성


입력: W₁ × H₁ × D₁  
하이퍼파라미터:  
- 필터 개수: K  
- 필터 크기: F  
- 스트라이드: S  
- 패딩: P  

출력 크기:  
W₂ = (W₁ - F + 2P) / S + 1  
H₂ = (H₁ - F + 2P) / S + 1  
D₂ = K

3) 풀링층

특징 맵의 크기를 줄여 계산량 감소 & 주요 특징만 남기기

종류:
- 최대 풀링(Max pooling): 영역의 최댓값 선택  
- 평균 풀링(Average pooling): 영역의 평균값 선택  

예: 2×2 영역에서 최대값 또는 평균값 선택하여 다운샘플링

- 최대 풀링 과정 예시

입력 2×2 영역에서:  
첫 번째 영역 → 최대값 3  
두 번째 영역 → 최대값 12  
세 번째 영역 → 최대값 3  
네 번째 영역 → 최대값 4  

- 평균 풀링 과정 예시

각 영역의 평균을 계산해 출력  
예: (3 + -1 + -3 + 1) / 4 = 0 등



4) 완전연결층 (Fully Connected Layer)

- 합성곱층과 풀링층을 거치며 차원이 축소된 특성 맵을 마지막으로 1차원 벡터로 펼친 후(flatten)
  완전연결층으로 전달한다.
- 이 층에서는 노드들이 모두 서로 연결되어 있으며, 분류 문제에서는 주로 소프트맥스(softmax) 함수를 사용한다.
- 소프트맥스 출력층에서는 각 클래스(레이블)에 대한 확률을 출력하며,
  가장 높은 확률을 가진 레이블이 최종 예측값이 된다.



### 5.1.3  1D, 2D, 3D 합성곱(CNN)
합성곱은 필터가 이동하는 방향/출력 형태에 따라 다음과 같이 나뉜다.
1D 합성곱 / 1×1 합성곱 / 2D 합성곱 / 3D 입력을 가진 2D 합성곱 / 3D 합성곱


 1) 1D 합성곱

- 필터가 시간축(가로 방향)으로만 이동하는 형태
- 입력이 [1, 1, 1, 1, 1] 이고, 필터가 [0.25, 0.5, 0.25] 라면
  출력은 [1, 1, 1] 형태가 된다.
- 주로 시계열 데이터나 1D 신호 데이터 처리 시 사용.


 2) 1 × 1 합성곱
- 3D 데이터(W, H, 채널)에 대해 필터(1, 1, L)를 적용.
- 출력은 (W, H).
- 채널 수를 조정하여 연산량(파라미터 수)을 크게 줄이는 효과.
- GoogLeNet의 Inception 구조에서 대표적으로 사용됨.


 3) 3D 입력을 가진 2D 합성곱
- 입력이 3D 형태 (예: 224×224×3, 112×112×32).
- 필터의 깊이(L)이 입력의 채널 수와 동일해야 함.
- 필터(k, k, L)를 적용 → 출력은 2D(W, H).
- CNN에서 가장 흔히 쓰이는 전형적인 합성곱.


4) 2D 합성곱
- 필터(k×k)가 가로(W), 세로(H)로 이동하는 일반적인 합성곱.
- 입력: (W, H)
- 필터: k×k
- 출력: (W, H)
- 이미지 처리에서 가장 많이 사용하는 형태.


5) 3D 합성곱
- 필터가 3개 방향으로 이동: (W, H, L)
- 입력: (W, H, L)
- 필터: (k, k, d)
- 출력: (W, H, L)
- 단, d < 입력 L 이어야 함.
- 주로 영상(video) 처리, 의료영상(CT·MRI) 등에서 사용됨.완전연결층 (Fully Connected Layer)

- 합성곱층과 풀링층을 거치며 차원이 축소된 특성 맵을 마지막으로 1차원 벡터로 펼친 후(flatten)
  완전연결층으로 전달한다.
- 이 층에서는 노드들이 모두 서로 연결되어 있으며, 분류 문제에서는 주로 소프트맥스(softmax) 함수를 사용한다.
- 소프트맥스 출력층에서는 각 클래스(레이블)에 대한 확률을 출력하며,
  가장 높은 확률을 가진 레이블이 최종 예측값이 된다.
